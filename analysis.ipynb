```python
import pandas as pd

df = pd.read_csv("data/keuangan.csv")
df.head()
```

```python
laba_per_tahun = df.groupby("Tahun")["Laba_Bersih"].sum()
laba_per_tahun
```

```python
# Analisis Laba Bersih per Perusahaan
# Analisis ini bertujuan untuk membandingkan total laba bersih masing-masing perushaan 
# yang menjadi sampel penelitian selama periode 2019â€“2022.
```

```python
laba_per_perusahaan = (
    df.groupby("Kode")["Laba_Bersih"]
      .sum()
      .sort_values(ascending=False)
)

laba_per_perusahaan
```

```python
plt.figure(figsize=(10, 6))
laba_per_perusahaan.plot(kind="bar")

plt.title("Total Laba Bersih per Perusahaan (2019â€“2022)")
plt.xlabel("Kode Perusahaan")
plt.ylabel("Total Laba Bersih")
plt.xticks(rotation=45)
plt.grid(axis="y")
plt.show()
```

```python
# Analisis Rasio Keuangan (Return on Assets / ROA)
# ROA digunakan untuk mengukur kemampuan perusahaan dalam menghasilkan
# laba dari total aset yang dimiliki.
```

```python
df["ROA"] = df["Laba_Bersih"] / df["Total_Aset"]
df.head()
```

```python
roa_perusahaan = (
    df.groupby("Kode")["ROA"]
      .mean()
      .sort_values(ascending=False)
)

roa_perusahaan
```

```python
plt.figure(figsize=(10, 6))
roa_perusahaan.plot(kind="bar")

plt.title("Rata-rata ROA per Perusahaan (2019â€“2022)")
plt.xlabel("Kode Perusahaan")
plt.ylabel("ROA")
plt.xticks(rotation=45)
plt.grid(axis="y")
plt.show()
```
