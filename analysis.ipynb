```python
df.groupby("Kode")["Laba_Bersih"].mean().sort_values(ascending=False)
```

```python
import matplotlib.pyplot as plt

df.groupby("Kode")["Laba_Bersih"].mean().sort_values().plot(kind="barh", figsize=(10,6))
plt.title("Rata-rata Laba Bersih per Perusahaan")
plt.xlabel("Laba Bersih (Rupiah)")
plt.ylabel("Kode Perusahaan")
plt.show()
```
